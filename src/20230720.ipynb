元の区間 PCM $A_k$

$\mathscr{A}_{ij}^- \subseteq A_{kij}$ ： $|\log ( \mathscr{A}_{ij}^- \cap A_{kij} )| / |\log \mathscr{A}_{ij}^-|$

$A_{kij} \subseteq \mathscr{A}_{ij}^+$ ： $|\log ( \mathscr{A}_{ij}^+ \cap A_{kij} ) | / |\log A_{kij}|$

各個人で整合化した二重区間 PCM $\tilde{A}_k$

$\mathscr{A}_{ij}^- \subseteq \tilde{A}_{kij}^-$ ： $|\log ( \mathscr{A}_{ij}^- \cap \tilde{A}_{kij}^- ) | / |\log \mathscr{A}_{ij}^-|$

$\tilde{A}_{kij}^+ \subseteq \mathscr{A}_{ij}^+$ ： $|\log ( \mathscr{A}_{ij}^+ \cap \tilde{A}_{kij}^+ ) | / |\log \tilde{A}_{kij}^+|$

手法 $l$ においてこれらの評価指標を集計した値を $\rho^{l-}, \rho^{l+}$ とする.

$i, j$ について算術平均, $k$ について幾何平均をとる.

```julia
function h(value, y_max, title)
    return histogram(
        value,
        bins=range(0, stop=1, length=21),
        legend=false,
        xlims=(-0.1, 1),
        ylims=(0, y_max),
        title=title
    )
end
```

```julia
function b(value, title)
    return boxplot(
        value,
        legend=false,
        xlims=(-0.1, 1),
        orientation=:horizontal,
        whisker_range=0,
        yticks=[],
        title=title
    )
end
```
