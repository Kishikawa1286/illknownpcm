```julia
function h(value, y_max, title)
    return histogram(
        value,
        bins=range(0, stop=1, length=21),
        legend=false,
        xlims=(-0.1, 1),
        ylims=(0, y_max),
        title=title
    )
end
```

```julia
function b(value, title)
    return boxplot(
        value,
        legend=false,
        xlims=(-0.1, 1),
        orientation=:horizontal,
        whisker_range=0,
        yticks=[],
        title=title
    )
end
```
